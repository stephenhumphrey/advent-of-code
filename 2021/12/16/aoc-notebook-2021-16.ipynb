```python
def peek(packet,bits):
    result = 0
    found_bits = 0
    cursor = 0
    
    while found_bits < bits and cursor < len(packet):
        chunk = packet[cursor] & split_mask
        cursor += 1
        
        need_bits = bits - found_bits
        if need_bits >= split_bits:
            result *= 2 ** split_bits
            result |= chunk
            found_bits += split_bits
        else:
            mini_chunk = chunk >> ( split_bits - need_bits )
            
            result *= 2 ** need_bits
            result |= mini_chunk
            found_bits += need_bits
            
    return result

test0 = 0b10110011100011110000111110000011
test1 = 0b11110000001111111000000011111111
test_packet = [test0,test1]
if verbose:
    for bits in range( split_bits * len(test_packet) + 1 ):
        value = peek( test_packet, bits )
        print( f"{bits:2} {split_formats[64].format(value)} = {value}" )
```

```python
def pop(packet,bits):
    #print(f"pop() bits = {bits}, packet = {packet}")
    result = peek(packet,bits)
    new_packet = packet
    
    bits_to_pop = bits
    while bits_to_pop >= split_bits:
        new_packet.pop(0)
        bits_to_pop -= split_bits
    
    if len(new_packet) > 0:
        for i in range(len(new_packet) - 1):
        
            new_packet[i] <<= bits_to_pop
            new_packet[i] &= split_mask            
            mini_packet = [new_packet[i + 1]]
            transfer_bits = peek(mini_packet,bits_to_pop)
            new_packet[i] |= transfer_bits
        
        new_packet[-1] <<= bits_to_pop
        new_packet[-1] &= split_mask
        # if new_packet[-1] == 0:
        #    new_packet.pop()
    
    return result,new_packet

print(test_packet)
if verbose:
    for bits in range( split_bits * len(test_packet) + 1 ):
        value,new_packet = pop(test_packet.copy(),bits)
        print( f"{bits:2} {split_formats[64].format(value)} {len(new_packet)}" )

destructive_test = test_packet.copy()
if verbose:
    for bits in range( split_bits * len(destructive_test) + 1 ):
        value,destructive_test = pop(destructive_test,1)
        print( f"{bits:2} {split_formats[8].format(value)} {len(destructive_test)} {split_format.format(destructive_test[0])} {split_format.format(destructive_test[-1])}" )
```

```python
def handle_literal(packet,version,type_id):
    
    bits_processed = 0
    
    value = 0
    while True:
        more,packet = pop(packet,1)
        bits_processed += 1
        
        chunk,packet = pop(packet,4)
        bits_processed += 4
        
        value <<= 4
        value |= chunk & 0xF
        
        if more == 0:
            break
        
    print(f"packet literal = {value}")
    
    return value,bits_processed
```
